### Import Library ###

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
```

## 1.1 Image Color Order with different libraries ##

```python
image = cv2.imread('./assets/images/groupImage.jpg')

image_B,image_G,image_R = cv2.split(image)

plt.figure(figsize=(15,10))

plt.subplot(2,4,1)
plt.imshow(image)
plt.title('BGR')

plt.subplot(2,4,2)
plt.imshow(image_B, cmap='gray')
plt.title('B')

plt.subplot(2,4,3)
plt.imshow(image_G, cmap='gray')
plt.title('G')

plt.subplot(2,4,4)
plt.imshow(image_R, cmap='gray')
plt.title('R')

rgbImage = image[:,:,::-1] # convert BGR to RGB

image_R,image_G,image_B = cv2.split(rgbImage)

plt.subplot(2,4,5)
plt.imshow(rgbImage)
plt.title('RGB')

plt.subplot(2,4,6)
plt.imshow(image_R, cmap='gray')
plt.title('R')

plt.subplot(2,4,7)
plt.imshow(image_G, cmap='gray')
plt.title('G')

plt.subplot(2,4,8)
plt.imshow(image_B, cmap='gray')
plt.title('B')

plt.figure(figsize=(2, 2))
plt.show()
```

## 1.3 Reduce Bit Depth using Quantization ##

```python
bit_depth = 2
Q_LEVEL = 2 ** bit_depth

def quantization(array_grayscale, q_level):

    array_copy = array_grayscale.copy()

    array_copy = np.floor((array_copy / 255) * q_level)

    return array_copy

image = cv2.imread('./assets/images/groupImage.jpg', 0)

print(image)

plt.figure(figsize=(10,10))
quantized_image = quantization(image, Q_LEVEL)
plot_image = plt.subplot(1,2, 1)
plt.imshow(image, cmap='gray')
plt.title('Image')

plot_quantized = plt.subplot(1,2,2)
plt.imshow(quantized_image, cmap='gray')
plt.title('quantized Image')
```

## 1.4 3D Image Surface ##

```python
img3d = cv2.imread('./assets/images/groupImage.jpg', 0)
print(img3d.shape)
img3d = cv2.resize(img3d, (100,100))
print(img3d.shape)
xx, yy = np.mgrid[0:img3d.shape[0], 0:img3d.shape[1]]

print(xx)
print(yy)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(xx, yy, img3d ,rstride=2,cstride=10, cmap='gray' ,linewidth=10)

plt.show()
```
